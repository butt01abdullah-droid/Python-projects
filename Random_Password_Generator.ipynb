```python
# Random Password Generator
import random,string

# Step 1:Define Password Generator function
def generate_password(length=12):
  if length < 4:
    raise ValueError("Password length should be at least 4 characters")

  # Character set for password
  uppercase= string.ascii_uppercase
  lowercase= string.ascii_lowercase
  digits= string.digits
  special_characters= string.punctuation

  # Ensure at least one of each character type
  password= [
      random.choice(uppercase),
      random.choice(lowercase),
      random.choice(digits),
      random.choice(special_characters)
  ]
  # Fill all the remaining with random choice of all sets
  all_char= uppercase+lowercase+digits+special_characters
  password+= random.choices(all_char, k=length-4)

  # Shuffle all to make it more random
  random.shuffle(password)

  # return the results
  return ''.join(password)

# Step 2: Interacting with user
try:
  length= int(input("Enter the password length (minimun 4 characters)"))
  password= generate_password(length)
  print("Generated Password:", password)
except ValueError as e:
  print(e)
```
