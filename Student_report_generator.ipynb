```python
# Student Report Generator
import csv

# Read data and Calculate Average
def process_student_data(input_file,output_file):
  try:
    with open(input_file, 'r') as file:
        reader = csv.DictReader(file)
        student_report=[]

        for row in reader:
          name = row['Name']
          roll_no = row['Roll no']
          math= int(row['Math'])
          physics = int(row['Physics'])
          chemistry = int(row['Chemistry'])
          computer = int(row['Computer'])
          average=round((math+physics+chemistry+computer)/4,2)
          status="Pass" if average>=40 else "Fail"

          student_report.append({
              'Name': name,
              'Roll no': roll_no,
              'math': math,
              'physics': physics,
              'chemistry': chemistry,
              'computer': computer,
              'average': average,
              'status': status
          })
    # Step 2: Write processed data to a new file
    with open(output_file, 'w', newline='') as file:
      fieldnames = ['Name', 'Roll no', 'math', 'physics', 'chemistry', 'computer', 'average', 'status']
      writer = csv.DictWriter(file, fieldnames=fieldnames)
      writer.writeheader()
      writer.writerows(student_report)
    print(f"Student report generated in {output_file}succesfully")
  except FileNotFoundError:
    print(f"File not found: {input_file}")
  except KeyError:
    print("Invalid column name in the input file.")
  except Exception as e:
    print(f"An error occurred: {e}")

# Step 3:Main program
input_file = 'student.csv'
output_file = 'student_report.csv'
process_student_data(input_file,output_file)
```
