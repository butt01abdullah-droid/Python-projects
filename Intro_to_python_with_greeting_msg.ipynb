```python
import random
secret_number= random.randint(1,10)
attempts=3
print("guess the number between 1 and 10")
while attempts>0:
  guess= int(input("guess a number"))
  if guess== secret_number:
    print("you win")
  elif guess < secret_number:
    print("too low")
  else:
    print("too high")
  attempts-=1
print("game over")


```
