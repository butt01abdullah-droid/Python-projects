```python
# Ingredients checker
# Step 1:Define the recipe ingredients
recipe_ingredients = {"flour", "sugar", "eggs", "butter"}

# Step 2: Get the user input for available ingeredients (assume data is coming in list)
available_ingredients = input("Enter the available ingredients (separated by commas): ").split(",")
available_ingredients= set(available_ingredients)

#Step 3: Compare ingredients
missing_ingredients = recipe_ingredients - available_ingredients
extra_ingredients = available_ingredients - recipe_ingredients

# Step 4: Display ingredients
print("\n----------Ingredients results------------")
if missing_ingredients:
    print("You are missing the following ingredients:", missing_ingredients)
else:
    print("You have all the ingredients needed.")

if extra_ingredients:
    print("You have extra ingredients:", extra_ingredients)


```
