```python
# Note Taking App

# Step 1:Define file name
FILE_NAME= "file.txt"

# Step 2: Display menu options
def display_menu():
  print("\n------Note Taking App-------\n")
  print("1. Add Note")
  print("2. View all Notes")
  print("3. Delete all Notes")
  print("4. Exit")

# Step 3:Add a new note
def add_note():
  note=input("Enter your note: ")
  with open(FILE_NAME,"a") as file:
    file.write(note + "\n")
  print("Note added successfully")

# Step 4:View all notes
def view_notes():
  try:
    with open(FILE_NAME,"r") as file:
      notes=file.read()
      if notes:
        print("\n---All Notes---\n")
        print(notes)
      else:
        print("No notes found")
  except FileNotFoundError:
    print("No notes found")
# Step 5: Delete all notes
def delete_notes():
  confirm=input("Are you sure you want to delete all notes? (yes/no): ")
  if confirm.lower()=="yes":
    with open(FILE_NAME,"w") as file:
      pass
    print("All notes deleted successfully")
  else:
    print("Deletion canceled")
# Step 6: Taking exit fron app
def exit_app():
  print("Exiting the app. Goodbye!")
  exit
# Step 7: Main program
while True:
  display_menu()
  choice=input("Enter your choice (1/2/3/4): ")
  if choice=="1":
    add_note()
  elif choice=="2":
    view_notes()
  elif choice=="3":
    delete_notes()
  elif choice=="4":
    exit_app()
  else:
    print("Invalid choice. Please try again")
```

```python
# Recipe Viewer App

# Step 1:Load Recipes from file
def load_recipes(file_path):
  try:
    with open(file_path,"r") as file:
      content=file.read()
      recipes=content.split("\n\n")
      recipe_dict={}
      for recipe in recipes:
        lines=recipe.split("\n")
        if len(lines)>=3:
          name=lines[0].strip()
          ingredients=lines[1].replace("Ingredients","").strip()
          instructions=lines[2].replace("Instructions","").strip()
          recipe_dict[name]={"ingredients":ingredients,"instructions":instructions}
      return recipe_dict
  except FileNotFoundError:
    return {}

# Step 2: Display Recipe Menu
def show_menu():
  print("\n-------Recipe Menu-------")
  print("1. View Recipes by Name")
  print("2. List all  Recipes")
  print("3. Exit")

# Step 3: Dispaly Recipe Details
def view_recipe(recipes):
  name=input("Enter the recipe name: ")
  if name in recipes:
    recipe=recipes[name]
    print(f"\n---{name}---")
    print(f"Ingredients: {recipe['ingredients']}")
    print(f"Instructions: {recipe['instructions']}")
  else:
    print("Recipe not found")

# Step 4: Main Program
recipe_file=("recipes.txt")
recipes=load_recipes(recipe_file)
while True:
  show_menu()
  choice=input("Enter your choice (1/2/3): ")
  if choice=="1":
    view_recipe(recipes)
  elif choice=="2":
    print("\n---list All Recipes---")
    for name,recipe in recipes.items():
      print(f"{name}: {recipe['ingredients']}")
  elif choice=="3":
    print("Exiting the app. Goodbye!")
    break
  else:
    print("Invalid choice. Please try again")
```

```python
# Daily Journal Logger

# Step 1: Define the Daily journal
journal_file='journalfile.txt'

# Step 2: Add an entry to journal
def add_entry():
  date=input("Enter the date (YYYY-MM-DD): ")
  entry=input("Enter your journal entry: ")
  with open(journal_file,"a") as file:
    file.write(f"{date}: {entry}\n")
  print("Entry added successfully")

# Step 3: View all entries
def view_entries():
  try:
    with open(journal_file,"r") as file:
      entries=file.read()
      if entries :
        print("\n---All Entries---")
        print(entries)
      else:
        print("No entries found")
  except FileNotFoundError:
    print("No entries found")

# Step 4: Search for an entry using date
def search_entry1():
  search_date=input("Enter the date to search (YYYY-MM-DD): ")
  try:
    with open(journal_file,"r") as file:
      for line in file:
        if search_date in line:
          print(line.strip())
          return
      print("Entry not found")
  except FileNotFoundError:
    print("No entries found")

# Step 5: Search for an entry using Keyword
def search_entry():
  search_keyword=input("Enter the keyword to search: ")
  try:
    with open(journal_file,"r") as file:
      for line in file:
        if search_keyword in line:
          print(line.strip())
  except FileNotFoundError:
    print("No entries found")

# Step 6: Main Menu
def show_menu():
  print("\n--------Daily Journal Logger-------")
  print("1. Add Entry")
  print("2. View All Entries")
  print("3. Search by Date")
  print("4. Search by Keyword")
  print("5. Exit")

# Step 7: Main program
while True:
  show_menu()
  choice=input("Enter your choice (1/2/3/4/5): ")
  if choice=="1":
    add_entry()
  elif choice=="2":
    view_entries()
  elif choice=="3":
    search_entry1()
  elif choice=="4":
    search_entry()
  elif choice=="5":
    print("Exiting the app. Goodbye!")
    break
  else:
    print("Invalid choice. Please try again enter between 1 to 5")
```
