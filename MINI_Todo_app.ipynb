```python
# Mini To-Do App
import json
import os

# File for storing tasks
Task_file = "my_tasks.json"

# Ensure the tasks file exist
if not os.path.exists(Task_file):
    with open(Task_file, 'w') as f:
        json.dump([], f)

# Step 1: Load tasks for json
def load_tasks():
    with open(Task_file, 'r') as f:
        return json.load(f)

# Step 2: Save tasks to json
def save_tasks(tasks):
    with open(Task_file, 'w') as f:
        json.dump(tasks, f, indent=4)

# Step 3: Add a new tasks
def add_task(): # Removed the 'task' parameter here, as it's not used/needed
    task_name = input("Enter a new task: ")
    tasks = load_tasks()
    tasks.append({"task": task_name , "status": "incomplete"})
    save_tasks(tasks)
    # Changed f-string message to reflect what was actually added
    print(f"Task '{task_name}' added successfully!")

# Step 4: View all tasks
def view_tasks():
    tasks = load_tasks()
    if not tasks:
        print("No tasks found.")
    else:
        print("\n------To Do List-------")
        for index, task in enumerate(tasks, start=1):
            print(f"{index}. {task['task']} - {task['status']}")

# Step 5: Search for any task
def search_task():
    search_term = input("Enter a search task name: ")
    tasks = load_tasks()
    found_tasks = [task for task in tasks if search_term.lower() in task['task'].lower()]
    if not found_tasks:
        print("No tasks found.")
    else:
        print("Tasks:")
        for index, task in enumerate(found_tasks, start=1):
            print(f"{index}. {task['task']} - {task['status']}")

# Step 6: Update Task status
def update_task_status():
    tasks= load_tasks()
    view_tasks()
    try:
        task_index = int(input("Enter the task number to update: "))
        if 1 <= task_index <= len(tasks):
            new_status = input("Enter the new status (complete/incomplete): ")
            tasks[task_index - 1]['status'] = new_status
            save_tasks(tasks)
            print("Task status updated successfully!")
        else:
            print("Invalid task number.")
    except ValueError:
        print("Invalid input. Please enter a valid task number.")

# Step 7: Delete a task
def delete_task():
    view_tasks()
    try:
        # Load tasks AFTER viewing them to ensure we have the most current list before deletion
        tasks = load_tasks()
        task_index = int(input("Enter the task number to delete: ")) - 1
        if 0 <= task_index < len(tasks):
            del tasks[task_index]
            save_tasks(tasks)
            print("Task deleted successfully!")
        else:
            print("Invalid task number.")
    except ValueError:
        print("Invalid input. Please enter a valid task number.")

# Step 8: Menu for user tasks - REMOVED the while True loop here
def display_menu():
    print("\nMenu:")
    print("1. Add a new task")
    print("2. View all tasks")
    print("3. Search for a task")
    print("4. Delete a task")
    print("5. Update task status")
    print("6. Exit")

# Step 9: Main program loop (This is the only loop needed)
while True:
    display_menu() # Calls the function to display options once
    choice = input("Enter your choice (1/2/3/4/5/6): ")
    if choice == '1':
        add_task()
    elif choice == '2':
        view_tasks()
    elif choice == '3':
        search_task()
    elif choice == '4':
        delete_task()
    elif choice == '5':
        update_task_status()
    elif choice == '6':
        print("Exiting the program.")
        break # Exits the main 'while True' loop
    else:
        print("Invalid choice. Please select a valid option.")
```
